# Plot greenspace metrics at multiple spatial scales

```python
import geopandas as gpd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## 1. Plot greenspace metrics at 100m

```python
# read GeoJSON containing greenspace metrics at 100m
gdf_100 = gpd.read_file("thesis_project/output/greenspace_metrics_100.json")
gdf_100.head()
```

```python
# read GeoJSON containing greenspace metrics at 100m within GM
gdf_100_gm = gpd.read_file("thesis_project/output/greenspace_metrics_100_gm.json")
gdf_100_gm.head()
```

### 1.1. NDVI vs. Greenspace%

```python
# plot NDVI vs. Greenspace%
sns.lmplot(x="NDVI_100", y="Greenspace%_100", data=gdf_100, 
          scatter_kws={"color": "green"}, line_kws={'color': 'lightgreen'})
```

#### 1.1.1. NDVI vs. Greenspace% within GM

```python
# plot NDVI vs. Greenspace% within GM
ndvi_green_100_plot = sns.lmplot(x="NDVI_100", y="Greenspace%_100", data=gdf_100_gm, 
          scatter_kws={"color": "green"}, line_kws={'color': 'lightgreen'})
```

```python
# save plot into png file
fig = ndvi_green_100_plot.fig
fig.savefig("thesis_project/visualizations/ndvi_green_100_plot.png") 
```

### 1.2. NDVI vs. Canopy%

```python
# plot NDVI vs. Canopy%
sns.lmplot(x="NDVI_100", y="Canopy%_100", data=gdf_100,
          scatter_kws={"color": "green"}, line_kws={'color': 'lightgreen'})
```

#### 1.2.1. NDVI vs. Canopy% within GM

```python
# plot NDVI vs. Canopy% only points within GM
ndvi_canopy_100_plot = sns.lmplot(x="NDVI_100", y="Canopy%_100", data=gdf_100_gm,
          scatter_kws={"color": "green"}, line_kws={'color': 'lightgreen'})
```

```python
# save plot into png file
fig = ndvi_canopy_100_plot.fig
fig.savefig("thesis_project/visualizations/ndvi_canopy_100_plot.png") 
```

### 1.3. Greenspace% vs. Canopy%

```python
# plot Greenspace% vs. Canopy%
sns.lmplot(x="Greenspace%_100", y="Canopy%_100", data=gdf_100,
          scatter_kws={"color": "green"}, line_kws={'color': 'lightgreen'})
```

#### 1.3.1. Greenspace% vs. Canopy% within GM

```python
# plot Greenspace% vs. Canopy% within GM
green_canopy_100_plot = sns.lmplot(x="Greenspace%_100", y="Canopy%_100", data=gdf_100_gm,
          scatter_kws={"color": "green"}, line_kws={'color': 'lightgreen'})
```

```python
# save plot into png file
fig = green_canopy_100_plot.fig
fig.savefig("thesis_project/visualizations/green_canopy_100_plot.png") 
```
