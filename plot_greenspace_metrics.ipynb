# Plot greenspace metrics at multiple spatial scales

```python
import geopandas as gpd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## 1. Plot greenspace metrics at 100m

```python
# read GeoJSON containing greenspace metrics at 100m within GM
gdf_100_gm = gpd.read_file("thesis_project/output/dataframes/greenspace_metrics_100_gm.json")
gdf_100_gm.head()
```

### 1.1. NDVI vs. Greenspace%

```python
# plot NDVI vs. Greenspace% within GM
ndvi_green_100_plot = sns.lmplot(x="NDVI_100", y="Greenspace%_100", data=gdf_100_gm, 
          scatter_kws={"color": "forestgreen"}, line_kws={'color': 'black'})
```

```python
# save plot into png file
fig = ndvi_green_100_plot.fig
fig.savefig("thesis_project/visualizations/eda/ndvi_green_100_plot.png") 
```

### 1.2. NDVI vs. Canopy%

```python
# plot NDVI vs. Canopy% only points within GM
ndvi_canopy_100_plot = sns.lmplot(x="NDVI_100", y="Canopy%_100", data=gdf_100_gm,
          scatter_kws={"color": "forestgreen"}, line_kws={'color': 'black'})
```

```python
# save plot into png file
fig = ndvi_canopy_100_plot.fig
fig.savefig("thesis_project/visualizations/eda/ndvi_canopy_100_plot.png") 
```

### 1.3. NDVI vs. Grass%

```python
# plot NDVI vs. Canopy% only points within GM
ndvi_grass_100_plot = sns.lmplot(x="NDVI_100", y="Grass%_100", data=gdf_100_gm,
          scatter_kws={"color": "forestgreen"}, line_kws={'color': 'black'})
```

```python
# save plot into png file
fig = ndvi_grass_100_plot.fig
fig.savefig("thesis_project/visualizations/eda/ndvi_grass_100_plot.png") 
```
